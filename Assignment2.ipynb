# Positive or Negative
# Write a program to check whether a given number is positive, negative, or zero.

```python
num = int(input("enter the number"))
if(num>0):
    print("positive")
elif(num<0):
    print("negative")
else:
    print("zero")
    
```

# Absolute Value
# Write a program to find the absolute value of a given number without using the abs() function.

```python
num = float(input("enter the number"))
if num<0:
    print(-num)
else:
    print(num)
```

# Leap Year Check
# Write a program to check whether a given year is a leap year or not. (Hint: A leap year is
# divisible by 4, but not by 100 unless also divisible by 400).

```python
year = int(input("enter the year"))
if((year%4==0 and year%100!=0) or (year%400==0)):
    print("year is leap year")
else:
    print("year is not leap year")
```

# Character Case Check
# Write a program to input a character and check whether it is an uppercase letter, lowercase letter, or not a letter.

```python
char = input("enter the character")
if 'A'<=char<'Z':
    print("uppercase")
elif 'a'<=char<='z':
    print("lowercase")
else:
    print("not a letter")
```

# Discount Calculator
# Write a program that calculates the discount:

```python
price = float(input("enter the price"))
discount_per = float(input("enter discount percentage"))
discount = (price*discount_per)/100
final_price = price-discount
print(f"Discount:{discount}")
print(f"Final Price:{final_price}")
```
